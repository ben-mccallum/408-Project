```python
from openttdlab import *
import pandas as pd
import plotly.express as px

results = run_experiments(
    openttd_version='13.4',
    opengfx_version='7.1',
    experiments=(
        {
            'seed': seed,
            'ais': (
                local_folder('proto', 'proto', ai_params=()),
                #bananas_ai('54524149', 'trAIns', ai_params=()),
            ),
            'days': 365 * 8 + 1,
        }
        for seed in range(15,35)
    ),
)

df = pd.DataFrame(
    {
        'seed': row['experiment']['seed'],
        'date': row['date'],
        'money': row['chunks']['PLYR']['0']['money'],
    }
    for row in results
)
df = df.pivot(index='date', columns='seed', values='money')
fig = px.line(df)
fig.show()
```
